## Use Neural Networks (NN) for temperature prediction


Now, we are trying to use NN to predict the surface air temperature (tas), based on the time series of CO2 & CH4 emissions

### 1. data preprocssing: prepare data for training & test sets

#### import data as training & test sets

Here we train NN using simulations from 3 historical and 3 future scenarios. <br/> Then we test the trained NN using the ssp245 scenario.

```python
# Training set
train_files    = ["historical", "ssp585", "ssp126", "ssp370","hist-aer","hist-GHG"]
X_train_xr, _  = prepare_predictor(train_files,train_path)
y_train_xr, _  = prepare_predictand(train_files,train_path)

# Test set
X_test_xr, _ = prepare_predictor('ssp245', data_path=test_path,time_reindex=False)
y_test_xr, _ = prepare_predictand('ssp245',data_path=test_path,time_reindex=False)
```

#### select relevant variables

predictors: CO2 & CH4 <br/>
predictand: tas

```python
X_train_df = pd.DataFrame({"CO2": X_train_xr["CO2"].data,
                           "CH4": X_train_xr["CH4"].data
                          }, index=X_train_xr["CO2"].coords['time'].data)

X_test_df  = pd.DataFrame({"CO2": X_test_xr["CO2"].data,
                           "CH4": X_test_xr["CH4"].data
                          }, index=X_test_xr["CO2"].coords['time'].data)


y_train_df = y_train_xr["tas"].stack(dim=["latitude", "longitude"])
y_train_df = pd.DataFrame(y_train_df.to_pandas())
```

This is how our predictors & predictand data look like:

Note that here we stack the original 2-D tas data into a single dimension (for the purpose of NN implementation)

```python
y_train_df.head()
```

#### Data normalization

Let's normalize the input predictors by there mean & std

```python
# Standardization
mean, std = X_train_df.mean(), X_train_df.std()

X_train_df   = (X_train_df - mean)/std
X_test_df    = (X_test_df - mean)/std

X_train = X_train_df.to_numpy()
y_train = y_train_df.to_numpy()
X_test = X_test_df.to_numpy()

print(X_train.shape,y_train.shape)
```

### 2. Define the NN structure

Now the data preprocessing has been finished! Let's define our NN structure, which is shown in the below schematic.

Here the NN we use has 3 hidden layers, and each hidden layer has 64 neurons. 

The NN input layer has 2 neurons, corresponding to CO2 & CH4 inputs respectively. 

The NN outputs are the global tas, each neuron of the output layer corresponds to each pixel. There are 13824 pixels in total (number of latitude: 96, number of longitude: 144)

![title](https://drive.google.com/uc?export=view&id=1wCIEX6GAN5h0_XOd23eUtdnkjpOJPpde)

Here are the hyperparameters for the NN training, such as the number of neurons per layer, learning rate, etc. Note that these hyperparameters here are for demonstration purposes only - they are not optimized.  

```python
# set hyperparameters
n_neuron       = 64
activation     = 'relu'
num_epochs     = 50
learning_rate  = 0.001
minibatch_size = 64
model_num      = 1
N_layers       = 2 # number of hidden layers
```

Here we compile 4 models with different loss functions, including mean squared error (mse), mean absolute error (mae), mean absolute percentage error (mape), and 50th quantile error (qloss).

Options for loss functions can be found <a href="https://keras.io/api/losses/">here</a>. You may also customize loss funciton, as shown in quantile loss below.

```python
# let us build a neural network from scratch

# using mse as loss function (mse: mean squared error)
model_mse = Sequential()
model_mse.add(Dense(n_neuron,  activation=activation, name="hidden_layer_1" ,input_shape=(X_train.shape[1],)))
for n in range(N_layers-1):
    model_mse.add(Dense(n_neuron,  activation=activation, name="hidden_layer_"+str(n+2))) 
model_mse.add(Dense(y_train.shape[1],  activation='linear', name="output_layer")) # the output layer
model_mse.compile(loss='mse',optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate))
model_mse.summary()
print('************************************************')


# using mae as loss function (mae: mean absolute error)
model_mae = Sequential()
model_mae.add(Dense(n_neuron,  activation=activation, name="hidden_layer_1" ,input_shape=(X_train.shape[1],)))
for n in range(N_layers-1):
    model_mae.add(Dense(n_neuron,  activation=activation, name="hidden_layer_"+str(n+2))) 
model_mae.add(Dense(y_train.shape[1],  activation='linear', name="output_layer")) # the output layer
model_mae.compile(loss='mse',optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate))
model_mae.summary()
print('************************************************')


# using mape as loss function (mape: mean absolute percentage error)
model_mape = Sequential()
model_mape.add(Dense(n_neuron,  activation=activation, name="hidden_layer_1" ,input_shape=(X_train.shape[1],)))
for n in range(N_layers-1):
    model_mape.add(Dense(n_neuron,  activation=activation, name="hidden_layer_"+str(n+2))) 
model_mape.add(Dense(y_train.shape[1],  activation='linear', name="output_layer")) # the output layer
model_mape.compile(loss='mean_absolute_percentage_error',optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate))
model_mape.summary()
print('************************************************')


# using quantile loss function 
def quantile_loss(y_true, y_pred):
    q = 0.5 # 50th quantile
    e = y_true-y_pred
    loss = tf.keras.backend.mean(tf.keras.backend.maximum(q*e, (q-1)*e))
    return loss
model_qloss = Sequential()
model_qloss.add(Dense(n_neuron,  activation=activation, name="hidden_layer_1" ,input_shape=(X_train.shape[1],)))
for n in range(N_layers-1):
    model_qloss.add(Dense(n_neuron,  activation=activation, name="hidden_layer_"+str(n+2))) 
model_qloss.add(Dense(y_train.shape[1],  activation='linear', name="output_layer")) # the output layer
model_qloss.compile(loss=quantile_loss,
                    optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate))
model_qloss.summary()
print('************************************************')
```

### 3. Train & save the NN model

Here we use `EarlyStopping` to avoid over-fitting

and we train 4 models using different loss functions

```python
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)

### train NN models
# ----------mse----------
history_mse = model_mse.fit(X_train, y_train, 
                            batch_size      = minibatch_size,
                            epochs          = num_epochs,
                            validation_split= 0.2, 
                            verbose         = 1,
                            callbacks       = [early_stop])
print('****************** end of mse ***********************')
# ----------mae----------
history_mae = model_mae.fit(X_train, y_train, 
                            batch_size      = minibatch_size,
                            epochs          = num_epochs,
                            validation_split= 0.2, 
                            verbose         = 1,
                            callbacks       = [early_stop])
print('****************** end of mae ***********************')
# ----------mape----------
history_mape = model_mape.fit(X_train, y_train, 
                              batch_size      = minibatch_size,
                              epochs          = num_epochs,
                              validation_split= 0.2, 
                              verbose         = 1,
                              callbacks       = [early_stop])
print('****************** end of mape ***********************')
# ----------quantile loss----------
history_qloss = model_qloss.fit(X_train, y_train, 
                                batch_size      = minibatch_size,
                                epochs          = num_epochs,
                                validation_split= 0.2, 
                                verbose         = 1,
                                callbacks       = [early_stop])
print('****************** end of qloss ***********************')
```

```python

plt.subplot(2, 2, 1)
plt.plot(history_mse.history['loss'])
plt.plot(history_mse.history['val_loss'])
plt.title('loss: Mean Squared Error', fontsize=12)
plt.ylabel('mse')
plt.xlabel('Epoch')
plt.legend(['Train loss', 'Val loss'], fontsize=10, loc='upper right')


plt.subplot(2, 2, 2)
plt.plot(history_mae.history['loss'])
plt.plot(history_mae.history['val_loss'])
plt.title('loss: Mean Absolute Error', fontsize=12)
plt.ylabel('mae')
plt.xlabel('Epoch')
plt.legend(['Train loss', 'Val loss'], fontsize=10, loc='upper right')


plt.subplot(2, 2, 3)
plt.plot(history_mape.history['loss'])
plt.plot(history_mape.history['val_loss'])
plt.title('loss: Mean Absolute Percentage Error', fontsize=10)
plt.ylabel('mape')
plt.xlabel('Epoch')
plt.legend(['Train loss', 'Val loss'], fontsize=10, loc='upper right')


plt.subplot(2, 2, 4)
plt.plot(history_qloss.history['loss'])
plt.plot(history_qloss.history['val_loss'])
plt.title('loss: Quantile Loss Function', fontsize=10)
plt.ylabel('quantile loss (50th)')
plt.xlabel('Epoch')
plt.legend(['Train loss', 'Val loss'], fontsize=10, loc='upper right')
plt.tight_layout()
plt.show()
```

Finally we check whether the ML model can capture the time series of global average temperature 

```python
def global_mean_std_plot(X,label,color,ax,var='tas'):
    weights  = np.cos(np.deg2rad(X.latitude))
    tas_mean = X[var].weighted(weights).mean(['latitude', 'longitude']).data
    tas_std  = X[var].weighted(weights).std(['latitude', 'longitude']).data
    
    x = X.time.data

    ax.plot(x, tas_mean, label=label,color=color,linewidth=2)
    ax.fill_between(x,tas_mean+tas_std,tas_mean-tas_std,facecolor=color,alpha=0.2)
```

```python
fig,ax = plt.subplots(figsize=(9,4))

global_mean_std_plot(y_test_xr,label='truth',ax=ax,color='tab:blue')
global_mean_std_plot(y_test_mse_pre,label='prediction (mse)',ax=ax,color='tab:orange')
global_mean_std_plot(y_test_mae_pre,label='prediction (mae)',ax=ax,color='tab:brown')
global_mean_std_plot(y_test_mape_pre,label='prediction (mape)',ax=ax,color='tab:gray')
global_mean_std_plot(y_test_qloss_pre,label='prediction (qloss)',ax=ax,color='tab:pink')

ax.set_xlabel('time')
ax.set_ylabel('global mean temperature (°C)')
ax.legend()
plt.tight_layout()
```
